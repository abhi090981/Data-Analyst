## Step.6 : Convert Date Format to a Consistent Type

```python
import datetime 
```

```python
def convert_to_consistent_format(date_str):
    for fmt in ("%d/%m/%Y"):
        try:
            date_obj = datetime.strptime(date_str, fmt)
            return date_obj.strftime("%Y-%m-%d")
        except ValueError:
            continue
    raise ValueError("Date format not recognized")
```
